```python
import pandas as pd

# Load the dataset
dataset_path = 'extended_movie_dataset.csv'  # Update with the actual file path
movie_df = pd.read_csv(dataset_path)

# Display the first few rows
print(movie_df.head())
```

```python
# Grouping by 'Director' and 'Genre' and counting the number of movies in each category
genre_director_counts = movie_df.groupby(['Director', 'Genre']).size().unstack()

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']

# Plotting the stacked bar chart
fig, ax = plt.subplots(figsize=(12, 6))
genre_director_counts.plot(kind='bar', stacked=True, ax=ax, color=colors)

# Set chart labels and title
ax.set_xlabel('Director')
ax.set_ylabel('Number of Movies')
ax.set_title('Distribution of Movie Genre by Director')

# Position the legend outside the plot for better visibility
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to make the chart readable
plt.tight_layout()

# Display the chart
plt.show()
```
